# ⏱️ Performance Testing - Tool Execution Times

## Test: Intelligence Query with Timing

```python
# Test query with detailed timing information
result = await run_query(
    "Peliculas donde actua Tom hanks",
    verbose=True
)

# Show summary
print("\n" + "="*80)
print("SUMMARY")
print("="*80)
if result:
    print(f"Selected Graph: {result.get('selected_graph', 'N/A')}")
    print(f"Status: {result.get('domain_graph_status', 'N/A')}")
    print(f"Schema Valid: {result.get('schema_valid', False)}")
    
    budget = result.get('budget_status', {})
    if budget:
        print(f"\nTotal Time: {budget.get('elapsed_time', 0):.2f}s")
        print(f"Total Tokens: {budget.get('total_tokens_used', 0)}")
        
        node_times = budget.get('node_execution_times', {})
        if node_times:
            slowest = max(node_times.items(), key=lambda x: x[1])
            print(f"Slowest Node: {slowest[0]} ({slowest[1]:.2f}s)")
print("="*80)
```

## Test: Pricing Query with Timing

```python
# Test pricing query with timing
result = await run_query(
    "¿Cuál es el precio actual de Netflix en Estados Unidos?",
    verbose=True
)

# Show timing breakdown
if result and result.get('budget_status'):
    budget = result['budget_status']
    node_times = budget.get('node_execution_times', {})
    
    if node_times:
        print("\n" + "="*80)
        print("TIMING BREAKDOWN")
        print("="*80)
        
        total = budget.get('elapsed_time', 0)
        print(f"Total: {total:.2f}s\n")
        
        for node, time_val in sorted(node_times.items(), key=lambda x: x[1], reverse=True):
            percentage = (time_val / total * 100) if total > 0 else 0
            bar = "█" * int(percentage / 2)
            print(f"{node:30s} {time_val:6.2f}s [{percentage:5.1f}%] {bar}")
        print("="*80)
```

## Test: Compare Multiple Queries

```python
# Compare timing across different query types
import pandas as pd

queries = [
    ("Intelligence", "Títulos exclusivos de Netflix en US"),
    ("Pricing", "Precio de Netflix en Estados Unidos"),
    ("Rankings", "Top 10 películas de Netflix en US"),
]

results_data = []

for query_type, question in queries:
    print(f"\n{'='*80}")
    print(f"Testing: {query_type}")
    print(f"{'='*80}")
    
    result = await run_query(question, verbose=False)
    
    if result and result.get('budget_status'):
        budget = result['budget_status']
        results_data.append({
            'Query Type': query_type,
            'Total Time (s)': budget.get('elapsed_time', 0),
            'Total Tokens': budget.get('total_tokens_used', 0),
            'Status': result.get('domain_graph_status', 'N/A'),
            'Graph': result.get('selected_graph', 'N/A')
        })

# Create comparison table
if results_data:
    df = pd.DataFrame(results_data)
    print("\n" + "="*80)
    print("PERFORMANCE COMPARISON")
    print("="*80)
    print(df.to_string(index=False))
    print("="*80)
    
    # Show statistics
    print(f"\nAverage Time: {df['Total Time (s)'].mean():.2f}s")
    print(f"Fastest: {df.loc[df['Total Time (s)'].idxmin(), 'Query Type']} ({df['Total Time (s)'].min():.2f}s)")
    print(f"Slowest: {df.loc[df['Total Time (s)'].idxmax(), 'Query Type']} ({df['Total Time (s)'].max():.2f}s)")
```
